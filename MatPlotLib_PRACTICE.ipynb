```python
#SUBPLOTS
x=pd.Series(np.arange(10))
y=x*2
y1=x*3
y2=x*4

plt.figure(figsize=(18,5))

plt.subplot(1,3,1)
plt.plot(x,y,color='m',marker='p') #p= pentagon shaped marker

plt.subplot(1,3,2)
plt.plot(x,y1,color='y',marker='*')

plt.subplot(1,3,3)
plt.plot(x,y2,color="g",marker='^')

plt.show()

# NOTE:b	blue ,g	green,r	red, c	cyan,m	magenta, y	yellow,k	black,w	white

```

```python
#A SINE WAVE

x = np.arange(1,100)
y = np.sin(x)

plt.figure(figsize=(5,2))
plt.plot(x, y)

plt.show()
```

```python
#LABELS ON THE GRAPH
x=pd.Series(np.arange(10))
y=x*2
y1=x*3
plt.figure(figsize=(18,5))

plt.subplot(1,2,1)
plt.plot(x,y,color='g',marker='^',label='label1')
plt.legend() #----------------------------------------- after each plot to get the labels,have to execute legend command

plt.subplot(1,2,2)
plt.plot(x,y,color='r',marker='*',label='label2')
plt.legend()

plt.show()
```

```python
#axhline,axvline   and 

x=pd.Series(np.arange(10))
y=x*2
plt.figure(figsize=(18,5))
plt.plot(x,y,'r^',label='label1')
plt.legend()

plt.axvline(2,0.1,0.8) # ymin n ymax is taken in terms of percentages ,here its 10% of (max y value) in tthe graph till 80% of (max y value)
plt.axhline(15) #not given xmin and xmax percentages

#Show coordinates or write text
plt.text(2,3, "blah")

#Save graph as an image 
# plt.saveflag('plot.png')

plt.show()

#SYNTAX:axvline(x=0, ymin=0, ymax=1, **kwargs)
#SYNTAX:axhline(y=0, xmin=0, xmax=1, **kwargs)
```

```python
#DIFFERENT GRAPH REPRESENTATIONS:

#1.Pie-chart -shows percentage representation of data

pi=[10,20,10,10,20] 
plt.pie(pi)

plt.show()

```

```python
#Labels on pi chart

p=[10,20,10,10,20] 
plt.pie(p,labels=p,explode=[0.5,0,0,0,0]) #explode is used the protrude a fraction of a pi chart.here we are protruding 1st value
plt.show()

plt.pie(p,labels=p,explode=[0.5,0,0.1,0,0])
plt.show()
```

```python
x=np.arange(10)
y1=np.random.randint(1,100,10)
y2=np.random.randint(1,100,10)

print(list(y1))
print(list(y2))

plt.figure(figsize=(18,5))

plt.subplot(1,3,1)
plt.bar(x,y1,color='m')

plt.subplot(1,3,2)
plt.bar(x,y2-y1,color='g')
plt.axhline(0)

plt.subplot(1,3,3)
plt.bar(x,y1,bottom=y2)

plt.show()
```

```python
x = [10,20,30,40,50,60]
y1 =[5,10,15,20,25,30]
y2 =[60,50,40,30,20,10]

print(y1,y2)

plt.figure(figsize=(16,5))
plt.subplot(1,3,1)
plt.bar(x,y1,color='g')
plt.bar(x,y2, bottom = y1,color='m',alpha=0.25) #it is used to increase transprency

#in this graph green is y1 values(y1 is at the bottom as it is mentioned above)
#pink is y2 values ,its stacked on top of each other


plt.subplot(1,3,2)
plt.bar(x,y2,color='k')
plt.bar(x,y1,bottom=y2,color='m')

#black is y2
#pink is y1


plt.show()
```

```python
x = np.arange(10)
y1 = np.random.randint(10,100,10)
y2 = np.random.randint(10,100,10)


plt.figure(figsize=(10,5))
plt.bar(x,y1,alpha=0.5, label=y1)
plt.bar(x,y2, alpha=0.5)
plt.bar(x, (y2-y1))
plt.axhline(0)

for l,h in zip(x , y1): #l is label ,h is height
    plt.text(l,h,str(h))

plt.show()
```

```python
#Note:can disply line plot on bar



#FILL BETWEEN

x = np.arange(10)
y1 = x *2
y2=x**2

plt.figure(figsize=(18,5))

plt.subplot(1,4,1)
plt.fill_between(x, y1,color='r')

plt.subplot(1,4,2)
plt.fill_between(x, y1,color='y', alpha = 0.4)
plt.fill_between(x, y2,color='c',alpha = 0.4)

plt.subplot(1,4,3)
plt.fill_between(x, y1,color='k',alpha = 0.4)
plt.fill_between(x, y2, y1,color='y',alpha = 0.4)

plt.subplot(1,4,4)
plt.fill_between(x, y1,color='g',alpha = 0.4)
plt.fill_between(x, y2+y1, y1,color='r', alpha = 0.4)

plt.show()
```

```python
# twinx
x = pd.Series(np.arange(100))
y = x*2
y1 = x**2

plt.figure(figsize=(15,5))

plt.plot(x,y, "r^", label="label1")
plt.twinx()
plt.plot(x,y1, "r-", label="label2")

plt.show()
```

```python
#HISTOGRAM

# Histogram Example

%matplotlib inline

H=[2,4,6,8,10,12,12,16,18,20,26,28,30,30,38,40,42,44,46,48,50]

plt.figure(figsize=(16,5))

plt.subplot(1,2,1)
plt.hist(H,color='m',bins=5,rwidth=0.8) #if bins are not mentioned.The default bin size is 10.

plt.subplot(1,2,2)
plt.hist(H,color='y',bins=[5,10,15,20,25,30,35,40,45,50],histtype='step')
```
