```python
#DATA CLEANSING

# s.isna()                	Whether there are null values existing      	Boolean
# df.dropna()             	Delete missing values                       	Can be applied on a series or a data frame
# df.fillna()             	Fill missing values with a certain value    	Can be applied on a series or a data frame
# df.drop_duplicates()    	Delete all duplicated values                	Can be applied on a series or a data frame
# df.drop()               	Delete columns or rows                      	Need to specify axis
# df.rename()             	Rename columns or rows                      	Need to specify axis
# df.reset_index()        	Convert index column to a data column       	A new index will be automatically
#                                                                           added using incremental integers
```

```python
#loc and iloc slicing
import pandas as pd
import numpy as np
df1 = pd.DataFrame(
    {
        "Name":   ["Gaurav", "Abhishek1", "Krishna", "Abhishek2", "Harshita", "Joey", "Shweta", "na-1" , "na-2"  , np.nan],
        "Age":    [ 0      , 1          , 2        , 2          , 1         , 2     , 3       , 1      ,  np.nan , 2     ],
        "Gender": [ 0      , 0          , 0        , 0          , 1         , 1     , 1       , np.nan ,  0      , 0     ]
    }
)
df2 = pd.DataFrame([("Gaurav",1), ("Abhiskek",2), ("Krishna",3), ("Abhishek2",4),('santorini',5),('hawaii',6)],
                    index=[0,11,12,7,3,2], 
                    columns=["Name", "Number"])
print(df1)
print()
print(df2)
```

```python
print(df1.loc[0])
print()
print(df1.iloc[0])
```

```python
print(df1.loc[:3]) #gets rows or columns with particular labels from the index(rows)
print()
print(df1.iloc[:3])# gets rows and columns with at particular positions in the index
print()
print(df2.loc[:3]) #prints everything untill the label is 3 in the index
print()
print(df2.iloc[:3])# prints 3 indexes-0,1,2 total 3
```

```python
#Aggregate operations

# count()              	Total number of items
# first(), last()      	First and last item
# mean(), median()     	Mean and median
# min(), max()         	Minimum and maximum
# std(), var()         	Standard deviation and variance
# mad()                	Mean absolute deviation
# prod()               	Product of all items
# sum()                	Sum of all items
```

```python
df = pd.DataFrame([[1, 2, 3],[4, 5, 6],[7, 8, 9],[np.nan, np.nan, np.nan]],columns=['A', 'B', 'C'])
print(df)
print()
print(df.agg(['sum', 'min']))# Aggregate these functions over the rows.
print()
print(df.agg({'A' : ['sum', 'min'], 'B' : ['min', 'max']})) # Different aggregations per column.

```

```python
print(df.agg("mean", axis="columns")) # Aggregate over the columns.
print()
print(df.agg("mean", axis=1))
print()
print(df.agg("mean", axis="rows"))# Aggregate over the rows.
print()
print(df.agg("mean", axis=0))
```

```python
print(df)
print()
print(df.min())
print()
print(df.min(axis=1))
print()
print(df.count(axis=1))
```

```python
#any and all

# Syntax:Series.any(self, axis=0, bool_only=None, skipna=True, level=None, **kwargs)

#ANY

df = pd.DataFrame({'c1': ['orange', 0], 'c2': ['grapes', 'berry']})
print(df)
print()
print(df.any())#------------------------------------------returns true if one of the value is true in the column or row
print()
print(df.any(axis=0))
print()
print(df.all()) #----------------------------------------returns false if one of the value is false in the column or row
print()
print(df.all(axis=1)) 
```

```python
#apply

# Syntax:s.apply(func, convert_dtype=True, args=())

tmp = pd.DataFrame(np.random.randint(10,20,20).reshape(5,4))
print(tmp)
print(tmp.apply(sum))
print()
print(tmp.apply(sum, axis=1))
print()
print(tmp.apply(lambda x : x**2, axis=1))
```
