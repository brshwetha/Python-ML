```python
#dist plot
sns.distplot(pokemon_data['Attack'])
plt.show()
#below shows a histogram,its called KDE curve
#at each bar a normal distribution curve is draw,overall it forms KDE
#y axis is frequency and x axis is attack points,exact frequency is shown in integers if there was just histograms,
#since KDE is a probability distribution curve it,y axis should be less than 1
```

```python
sns.distplot(pokemon_data['Attack'],color='pink')
plt.show()
```

```python
#Remove KDE,keep histogram
hist=sns.distplot(pokemon_data['Attack'],color='green',kde=False)
hist.set_title("Attack capability")
hist.set_xlabel("Attack point")
hist.set_ylabel("Frequency")
plt.show()
```

```python
#Remove hist,display KDE
hist=sns.distplot(pokemon_data['Attack'],color='green',hist=False)
hist.set_title("KERNEL DENSITY PLOT")
hist.set_xlabel("Attack point")
hist.set_ylabel("unit probability")
plt.show()
```

```python
#Rugged plot
hist=sns.distplot(pokemon_data['Attack'],color='maroon',rug=True)
hist.set_title("Attack capability with KDE and rug plot")
hist.set_xlabel("Attack point")
hist.set_ylabel("unit probability")
plt.show()
```

```python
#Categorical data plot

#Import tips dataset -available in sns library

tips=sns.load_dataset('tips')
print(tips)
```

```python
sns.set(style='whitegrid',color_codes=True) #using white board plotting
```

```python
#ONLY FOR CATEGORICAL DATA

#1.STRIP PLOT

sns.stripplot(x='day',y='tip',data=tips,size=5) #days- thur to sunday so--- days are categorical data
plt.show()                                      # x values are categorical,one of them should be categorical !



plt.subplots(figsize=(10,6))
sns.stripplot(x='day',y='tip',data=tips,size=6) 
plt.show()

plt.subplots(figsize=(10,6))
sns.stripplot(x='day',y='tip',data=tips,size=6,jitter=True)#jitter reduces the clumsiness,reduces the coincidence of the points 
plt.show()
```

```python
#2.SWARM PLOT - to reduce overlapping of points,for better representation

plt.subplots(figsize=(10,6))
sns.swarmplot(x='day',y='tip',data=tips) 
plt.show()

```

```python
plt.subplots(figsize=(10,6))
sns.swarmplot(x='day',y='tip',data=tips,size=8,hue='size')
plt.show()

```

```python
#horizontal :exchange x and y attributes

plt.subplots(figsize=(10,6))
sns.swarmplot(x='tip',y='day',size=8,data=tips) 
plt.show()

```

```python
#3.Box plot

plt.subplots(figsize=(15,5))
sns.boxplot(x='day',y='tip',data=tips) 
plt.show()
```

```python
#4.Violin plot

plt.subplots(figsize=(15,5))
sns.violinplot(x='day',y='tip',data=tips) 
plt.show()

#left and the right curved surfaces indicate kde plot
#Eg:more ppl are tipping 2$ on thursday so density around that range is high which is denoted by kde plot
```

```python
plt.subplots(figsize=(15,5))
sns.violinplot(x='day',y='tip',data=tips,bw=.1,scale='count') #bandwidth(bw): gives more precised distribution of points
plt.show()
```

```python
#SWARM PLOT WITH VIOLIN PLOT

plt.subplots(figsize=(15,5))
sns.violinplot(x='day',y='tip',data=tips,inner=None) 
sns.swarmplot(x='day',y='tip',data=tips,color='black') 
plt.show()
```

```python
f,ax=plt.subplots(figsize=(15,5))
sns.violinplot(x='day',y='tip',data=tips,inner='stick',palatte='Set3') #black lines because of inner=stick
plt.show()
```

```python
#Point plot

f,ax=plt.subplots(figsize=(15,5))
sns.pointplot(x='day',y='tip',data=tips)
plt.show()

#longer the line higher is the confidence interval
```

```python
#Factor plot function is renamed to catplot

sns.factorplot(x='day',y='tip',data=tips,hue='size',height=3,aspect=2,kind='bar')#kind is replaced by strip in the new version
plt.show()

#black lines are outliers 
```

```python
#BOX

sns.factorplot(x='day',y='tip',data=tips,col='size',height=5,aspect=2,kind='box')
plt.show()
```

```python
#SWARM

sns.factorplot(x='day',y='tip',data=tips,col='size',height=5,aspect=2,kind='swarm')
plt.show()
```
