```python
# Ignore the first 20 rows

# Solution:Treat the first 20 rows as header rows so that excel file will be read from 21st line

df=pd.read_excel('Australia.xlsx',sheet_name='Australia by Residence',header=20)
print(df)
#previously there were 468 rows now there are 448 rows,20 rows are considered as header
```

```python
# Drop
# Delete all Data where the Type = "Emigrants" or Type is not "Immigrants" (Row Deletion)
```

```python
df[df.Type!="Emigrants"] #shows the number of rows that do not have emmigrants 
```

```python
df[df.Type=="Emigrants"]#shows the number of rows having emigrants in the column type
```

```python
A=df[df.Type=='Emigrants']
print(A) #shows empty data frame because emmigrants is deleted and inplace option is used.
print()
df.drop(A.index,inplace=True) #it drops all the 224 rows having emigrants 
print(df) 

#first get the indexes of emigrants
#drop the emigrants rows 
```

```python
# Find columns for which more than 50% data is missing and drop them (Column Deletion)

#SYNTAX:DataFrame.drop(self, labels=None, axis=0, index=None, columns=None, level=None, inplace=False, errors='raise')

#THERE IS NO THRESH IN DROP HENCE USE DROPNA 

df.dropna(axis='columns',thresh=df.shape[0]*0.50,inplace=True)
print(df)
```

```python
# Drop Unnessary Columns: Type, Coverage, Area, Reg, Dev (Column deletion) Your columns should now look like "OdName,AreaName,RegName,DevName,1980 ... 2008

#DEL AND POP IS USED TO DELETE ONLY A SINGLE COLUMN ,TO DROP MULTIPLE COLUMNS USE DROP

df.pop('Type')
df.pop('Coverage') #displays the contents of coverage and deletes the column
```

```python
del df['AREA'] #does not display the contents but deletes the column
```

```python
#Drop multiple columns use drop()

df.drop(['DEV','REG'],axis='columns',inplace=True)
```

```python
print(df.columns)
print(len(df.columns))
print(len(df.index)) #ALWAYS USE THE WORD INDEX WHEN U WANT INFO RELATED TO ROWS
```

```python
# Rename

# Rename columns :OdName -> Country; RegName -> Region
df.rename(columns={'OdName':'Country'}) #renames a single column

#RENAME multiple columns

df.rename(columns={'OdName':'Country','RegName':'Region'},inplace='True')
print(df.columns)
```

```python
print(df.dtypes) #last 5 columns object data type
```

```python
import numpy as np
df.replace('..',np.nan,inplace=True) #all the vaues from 2009 to 2013 had '..' as their values.Its replaced with NAN 
```

```python
# Find Total number of immigrants for all years, and create a new column Total in data Frame

df['total'] = df.iloc[:,4:].sum(axis =1) #rows all rows ,columns 4th and the rest after 4th column
print(len(df.columns))
print()
print(df.total)
```

```python
#NOTE:INSERT SYNTAX: DataFrameName.insert(loc, column, value, allow_duplicates = False)
# df.insert(len(df.columns),'Total', list(a) , True) 
```

```python
# Head Tail Sort

# Sort Data by the new column Total in descending order and analyze the fist few rows. (get first few rows using head)

First_few_rows=(df.sort_values(by='total',ascending=False))

print(First_few_rows.head(2))
```

```python
# find something weird in the dataset ? Delete that row 'world'

#NOTE:HAVE TO KNOW THE COLUMN NAME OF WORLD

Weird=df[df.AreaName=='World'] #World is present in 2 rows in the AreaName column
print(Weird)
Weird.drop(Weird.index,inplace=True)
print()
print()
print(Weird)
```

```python
# Analyze last few rows also in the sorted result (Don't sort inplace)

print(df.tail(5))
print()
print(First_few_rows.tail(5))
```

```python
# Check if these have all unique values or not:

# countries -----------to check if every row in country has a unique value or not
print(df.Country.unique()) 
print()

if len(df.Country.unique())==len(df.Country):
    print('unique')
else:
    print('not unique')
```

```python
#continents (AreaName) -------------to check if every row in continent has a unique value or not
print(df.AreaName.unique())
print()
if len(df.AreaName.unique())==len(df.AreaName):
    print('unique')
else:
    print('not unique')
    
```

```python
# Find count of above countries
len(df.Country.unique())
print()

#OR

a=(df.Country.unique())
print(a.shape)
```

```python
# country continent pairs
len(df.groupby(by =['AreaName','Country']).groups) == df.shape[0] 
#if values of country continent pairs are unique ,if they are unique it will be equal to the number of rows in the dataframe
```

```python
# Find countries for which no of immigrants for all years is 0

immigrants_all_countries=df.Country[df.total ==0]
print(immigrants_all_countries)
print(immigrants_all_countries.count())
```

```python
# Find countries for which no of immigrants is 0 for any year
df.Country[df[df.columns[4:]].apply(lambda x:x==0, axis =0).any(axis=1)]
```
